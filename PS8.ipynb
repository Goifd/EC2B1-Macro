## Question 1
Weyl discrete Solow model:
$$
Y = AK^\alpha L^{\alpha-1}
$$

$Y = C+I$ and $I = S$ so $Y = C + S -> S = \gamma Y$
$$K'= (1-\delta)K+I$$

in steady state $K'=K$ so $K=(1-\delta)K+\gamma Y$ substituting in $Y$, $K=(1-\delta)K+\gamma AK^\alpha L^{\alpha-1}$ so $\delta K = \gamma AK^\alpha L^{\alpha-1}$
which means that at steady state $$K^{\alpha-1} = \frac{\delta}{\gamma A L^{\alpha-1}}$$
for $A = L = 1$
$$K = k = \left(\frac{\delta}{\gamma}\right)^{\frac{1}{\alpha-1}}$$ 


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
delta = 0.1
alpha = 0.3
gamma = 0.2
A = 1
L = 1
n = 0
g = 0

print("k is: ", (delta/gamma)**(1/(alpha-1)))
```

Now working with the per capita expressions and $A=L=1$, and $n=g=0$:

$$y_t = k_t^\alpha$$
$$k_{t+1}=(1-\delta)k_t+\gamma y_t$$

```python
k_arr = [1.0]
y_arr = [1.0]
t = np.arange(0,1001)
for i in range(1000):
    k = (1-delta)*k_arr[i]+gamma*y_arr[i]
    k_arr.append(k)
    y = k**alpha
    y_arr.append(y)
print("k steady state is: ", k_arr[-1])
```

```python
plt.plot(t[:100],k_arr[:100],label="k")
plt.plot(t[:100],y_arr[:100],label="y")
#plt.plot(t[:100],delta*np.array(k_arr[:100]),label="delta*k")
#plt.plot(t[:100],gamma*np.array(y_arr[:100]),label="gamma*y")
plt.legend()
```
